```python
def numsin(x,i):
    j=1
    a=x
    resp=0
    while(j<=i):
        j+=1
        b=(-x**2)/((2*j-1)*(2*j-2))*a
        resp+=a
        a=b
        
    return resp
```

```python
figure(figsize=(15,8))
x=linspace(0,35,1000)
plot(x,numsin(x,10000))
ylabel("numsin(x)",size=14)
xlabel("x",size=14)
```

```python
i=0
r=[]
r2=[]
while(i<80):
    j=0
    resp=True
    while(resp):
        a=numsin(0.45*i,j)
        if (abs(a-sin(0.45*i)))<0.01:
           resp = False
           r.append(j+1)
           r2.append(0.45*i)
        j+=1
    i+=1    
i2=0
print("Ángulo (rad)     Iteraciones necesarias")
print()
while(i2<len(r)):
    print ("%.2f" %r2[i2],"            ", r[i2])
    i2+=1
```

```python
figure(figsize=(10,10))
scatter(r2,r)
ylabel("Cantidad de iteraciones",size=14)
xlabel("Ángulo",size=14)
xlim(0,40)
ylim(0,55)
```
