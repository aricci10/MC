```python
%pylab inline
from scipy.integrate import odeint
from scipy.integrate import ode
from mpl_toolkits.mplot3d import Axes3D
```

```python
a = b = 0.1
c = 14
def derivadas(Y,t):
    global c
    return array([-Y[1]-Y[2],Y[0]+a*Y[1],b+Y[2]*(Y[0]-c)])
```

```python
tiempos=linspace(0,600,100000)
sol=odeint(derivadas,[1.,1.,1.],tiempos,printmessg=True)
```

```python
figure(figsize(10,10))
gca(projection='3d')
plot(sol[:,0],sol[:,1],sol[:,2])
```
