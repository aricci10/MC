```python
#Se define la función Fourier, la cual realiza n iteraciones a la suma de fourier de la onda cuadrada.
def fourier(x,n):
    i=0
    resp=0
    while(i<n):
        a=(1/(2*i+1))*sin(x*(2*i+1))
        resp+=a
        i+=1
    return resp

#Se define la función encargada de dar máximos de un arreglo que ingresa por parámetro
def darMaximo(arr):
    i=0
    resp=0
    while(i<len(arr)):
        
        if arr[i]>resp:
            resp=arr[i]
            
        i+=1
    return resp
#Se define la función encargada de dar mínimos de un arreglo que ingresa por parámetro
def darMinimo(arr):
    i=0
    resp=1000000
    while(i<len(arr)):
        
        if arr[i]<resp:
            resp=arr[i]
            
        i+=1
    return resp      
```

```python
x=linspace(0,2*pi,100000)
j=1
respj=[]
respG=[]
#Se corren 40 iteraciones para diferentes para hallar 40 valores de la constante
while(j<40):
    max=darMaximo(fourier(x,j))
    min=darMinimo(fourier(x,j))
    a=max-min
    respj.append(j)
    respG.append(a)
    j+=1
print ("El valor de la constante de Wilbraham-Gibbs es:", respG[len(respG)-1])
```

```python
#Se imprime la figura creada. 
figure(figsize(10,10))
scatter(respj,respG,label="Datos")
linea=[-10,50]
linea2=[respG[len(respG)-1],respG[len(respG)-1]]
plot(linea,linea2,c='r', label="Valor estimado")
xlabel('Iteraciones realizadas',size=12)
ylabel('Valor de la constante',size=12)
title('Constante de Wilbraham-Gibbs',size=18)
xlim(-10,50)
legend()
```
