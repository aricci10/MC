```python
%pylab inline
from scipy import interpolate
```

```python
#Se importan los datos como una matriz.
data=genfromtxt("Datos.csv", delimiter=",")
#Se establece el radio terrestre.
Radio=6371000
print(data)
```

```python
#Se crea una matriz de ceros donde ir'an las coordenadas cartesianas. El formato es es x,y,z,t.
dataCart=zeros((1863,4),dtype='float')
#Se realiza el cambio de coordenadas.
dataCart[:,0]=Radio*sin(radians(90-data[:,1]))*cos(radians(data[:,0]))
dataCart[:,1]=Radio*sin(radians(90-data[:,1]))*sin(radians(data[:,0]))
dataCart[:,2]=Radio*cos(radians(90-data[:,1]))
#Se normalizan el tiempo y el espacio, tal que arranquen en 0,0,0,0
dataCart[:,3]=(data[:,3]-11)*3600+(data[:,4]-21)*60+data[:,5]-4
dataCart[:,0]-=dataCart[0,0]
dataCart[:,1]-=dataCart[0,1]
dataCart[:,2]-=dataCart[0,2]
print (dataCart)
```

```python
def dot(x1,y1,z1,x2,y2,z2):
    return x1*x2+y1*y2+z1*z2
```

```python
#Se crea una matriz, en donde se colocan las coordenadas oriente-norte de de Bogotá. El formato es N,O,t
dataBogo=zeros((1863,3),dtype='float')
dataBogo[:,0]=-1*dot(dataCart[:,0],dataCart[:,1],dataCart[:,2],-sin(radians(90-4.5981)),cos(radians(90-4.5981)),0)
dataBogo[:,1]=dot(dataCart[:,0],dataCart[:,1],dataCart[:,2],cos(radians(90-4.5981))*cos(radians(-74.0758)),sin(radians(90-4.5981))*cos(radians(-74.0758)),-sin(radians(-74.0758)))
dataBogo[:,2]=dataCart[:,3]
print(dataBogo)
figure(figsize(10,10))
#Se realiza una gráfica del recorrido.
plot(dataBogo[:,0],dataBogo[:,1])
ylabel("Posición Norte (m)")
xlabel("Posición Oriente (m)")
title("Un paseo en el parque")
```

```python
#Se define una función encargada de realizar la integral numérica. A esta le entra un arreglo de posiciones y tiempo, y devuelve el tiempo y la distancia..
def Integral(arr):
    datX=arr[:,0]
    datY=arr[:,1]
    tiem=arr[:,2]
    datXdesp=roll(datX,1)
    datYdesp=roll(datY,1)
    datXf=datX-datXdesp
    datYf=datY-datYdesp
    datXf[0]=0
    datYf[0]=0
    distU=sqrt(datXf**2+datYf**2)
    dist=cumsum(distU)
    return dist,tiem
#Se ejecuta la función con los datos de Bogotá. y se grafican
distancia,t=(Integral(dataBogo))
figure(figsize(12,12))
plot(t,distancia) 
ylabel("Distancia(m)")
xlabel("Tiempo(s)")
title("Un paseo en el parque",size=16)
```

```python
#Ahora, se define una nueva función que se encarga de entregar el tiempo, la posición y la disntancia más próxima para una distancia dada.
def TiempoParaDist(dista):
    i=0
    temp=True
    resp=0
    while(temp): 
        if(distancia[i]>=dista):
            temp=False
            resp=i    
        i+=1
    distProx=distancia[i]
    lat=data[resp,0]
    lon=data[resp,1]
    tiempoo=dataBogo[resp,2]
    return(tiempoo,distProx,lat,lon)
```

```python
#Se corre la función creada anteriormente y se imprimen sus respuestas.
i=1000
while(i<9001):
    a,b,c,d=TiempoParaDist(i)
    print("")
    print("Para",i,"metros: ")
    print("")
    print("La siguiente distancia más próxima es:", b,"metros")
    print("")
    print("El tiempo necesario fue de:",a,"segundos")
    print("")
    print("Las coordenadas en dicha posición son (Longitud, Latitud):",c,d)
    print("")
    i+=1000
```

```python
#Ahora se interpolan los datos de distancia.
distanciaInterpolada=interpolate.interp1d(t,distancia)
```

```python
#Se generan nuevos datos de distancia separados uniformemente.
tiempoUniforme=linspace(0,t[len(t)-1],len(t)-1)
scatter(tiempoUniforme,distanciaInterpolada(tiempoUniforme))
distanciasUniformes=distanciaInterpolada(tiempoUniforme)
```

```python
#Se obtienen las derivadas, usando la intepolación recién creada.
h=(t[len(t)-1])/(len(t)-1)
DistanciaAdelante=roll(distanciasUniformes,-1)
Diferencia=(DistanciaAdelante-distanciasUniformes)
Diferencia[len(Diferencia)-1]=0
Derivada=Diferencia/h
```

```python
scatter(tiempoUniforme,Derivada)
ylabel("Rapidez (m/s)")
xlabel("Tiempo (s)")
title("Un paseo en el parque (Rapidez)")
ylim(0,25)
xlim(0,6000)
```

```python
#Ahora se sacan nuevas derivadas, disminuyendo el "Step size" 10 veces, aumentado 10 veces la cantidad de datos y aplicando diferencia central como método de derivación
tiempoUniforme2=linspace(0,t[len(t)-1],10*(len(t)-1))
h2=(t[len(t)-1])/(10*(len(t)-1))
distanciasUniformes2=distanciaInterpolada(tiempoUniforme2)
DistanciaAdelante2=roll(distanciasUniformes2,-1)
DistanciaAtras2=roll(distanciasUniformes2,1)
derivadaFinal=(DistanciaAdelante2-DistanciaAtras2)/2*h2
derivadaFinal[0]=0
derivadaFinal[len(derivadaFinal)-1]=0
```

```python
#Se grafica la nueva rapidez
scatter(tiempoUniforme2,derivadaFinal)
ylabel("Rapidez (m/s)")
xlabel("Tiempo (s)")
title("Un paseo por el parque (Rapidez final)")
ylim(0,3)
xlim(0,6000)
```
