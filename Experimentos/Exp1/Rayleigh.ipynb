El objetivo de este informe es observar si es posible modelar la distribución de velocidades, mediante una distribuciónde Rayleigh. Para ello se miden las posiciones y velocidades en dos dimensiones de 1600 partículas durante un tiempo, posteriormente mediante el uso de histogramas, se detemrina la cantidad de veces que se repite la rapidez (norma del vector velocidad). Finalmente se compara con el modelo teórico. 


### Procesamiento de datos y análisis

```python
#Se inicia por almacenar los datos experimentales en "data"
data=genfromtxt("Rayleigh.csv",delimiter=",")
```

```python
#Se separan los datos por columnas, siendo posición x, posición y, velocidad v y velocidad y, respectivamente cada columna
x=data[:,0]
y=data[:,1]
vx=data[:,2]
vy=data[:,3]
#Se toman los últimos datos de la simulación
ultimosx=x[-1601:-1]
ultimosy=y[-1601:-1]
ultimosvx=vx[-1601:-1]
ultimosvy=vy[-1601:-1]
```

```python
#Se crea un ciclo para sacar la norma del vector velocidad y la norma del vector posición. Se almacenan en dos arreglos.
i=0
rapidez=[]
pos=[]
while(i<1600):
    x=sqrt(ultimosvx[i]**2+ultimosvy[i]**2)
    z=sqrt(ultimosx[i]**2+ultimosy[i]**2)
    rapidez.append(x)
    pos.append(z)
    i+=1
```

```python
#Se realiza el histograma de los datos obtenidos.
figure(figsize=[10,10])
dataHist=hist(rapidez, bins=50, normed =True)
suptitle('Distribución de velocidades', fontsize=20)
xlabel('Velocidad', fontsize=18)
ylabel('', fontsize=16)
show()
```

```python
#Se define la función de distribución de Rayleigh
def Ray(x,a):
    return(x/a**2)*e**(-x**2/2*a**2)
```

```python
#Se obtienen los datos del histograma
Y=dataHist[0]
X=delete(dataHist[1],-1)
```

```python
#Se realiza el ajuste de las velocidades, usando la función Ray. El valor obtenido equivale al parámetro "a" de la función
fitpars,covmat=curve_fit(Ray,X,Y)

print(fitpars)
```

```python
#Ahora se realiza una gráfica que incluye los datos experimentales, el modelo experimental y el modelo teórico. En el módelo teórico "a" vale 1.
figure(figsize=[10,10])
x=linspace(0,4,1000)
plot(x,Ray(x,fitpars),c='g', label='Modelo experimental')
plot(x,Ray(x,1),c='b',label='Modelo teórico')
hist(rapidez, bins=50, normed =True, label='Datos obtenidos')
suptitle('Distribución de velocidades', fontsize=20)
xlabel('Velocidad', fontsize=18)
legend()
ylabel('', fontsize=16)
show()
```

```python
#Ahora se realiza la distribución de posición
figure(figsize=[10,10])
suptitle('Distribución de Posición', fontsize=20)
xlabel('Posición', fontsize=18)
dataHist2=hist(pos, bins=70, normed =True, label='Datos obtenidos')
show()
```

```python
#Se calcula el error:

error=(1-fitpars)*100
print(error)
```
